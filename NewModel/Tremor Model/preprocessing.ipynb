```python
main_folder = 'FullDataSet_PD-BioStampRC21'
```

```python
def get_folder_and_dataset(id_no):
    id_string = f'{id_no:03d}'
    folder_name = id_string
    data_name = 'rh_ID' + id_string + 'Accel.csv'

    return main_folder+'//'+folder_name+'//'+data_name

data_name = get_folder_and_dataset(63)
read_csv(data_name)
```

```python
def get_data_windows(id,n,window_size):
    data_name = get_folder_and_dataset(id)
    data = read_csv(data_name).to_numpy()
    print(data)
    top = np.max(data.shape)
    m_rand = np.random.randint(0,top,size=n)
    windows_list = []
    for i in m_rand:
        window = data[i:i+window_size,1:]
        window = np.sqrt(np.sum(np.pow(window,2),axis=1))
        windows_list.append(window)

    return np.array(windows_list)

get_data_windows(5,100,50).shape
```
